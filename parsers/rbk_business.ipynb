```python
id = 0
hrefs = []
items = []
for item in range(10):
    url = 'https://www.rbc.ru/v10/ajax/get-news-by-filters/?category=business&offset='+ str(id) +'&limit=12'
    req = requests.get(url, headers=headers)
    src = req.text

    soup = BeautifulSoup(src, 'lxml')  

    postHref = soup.find("a").get("href")
    items.append(
        {
            "id": id,
            "href": postHref
        }
    )
    id += 1
    print(postHref +'completed')
    hrefs.append(postHref)
    print(id)
```

```python

with open('rbk_economy_hrefs.json', 'w') as file:
    json.dump(items, file, indent=4, ensure_ascii=False)
```

```python
bigData = []
id = 0
for href in hrefs:
    data =[]
    url = href[2:-2]
    print(url)
    req = requests.get(url, headers=headers)
    src = req.text
    soup = BeautifulSoup(src, 'lxml')  

    title = soup.find(class_="article__header__title-in js-slide-title")

    title_text = title.text

    data.append(url)
    data.append(title_text)
    print(data)
    print(id)
    id+=1
    

    bigData.append(data)
```
